| Purpose    | Background <br> (`background-color`) | Text <br> (`color`) | Notes / Feeling |

| **Task**      | `#9370DB` (medium purple) | `#1A0033` (very dark purple) | Purple gives a "priority / important" feeling |
| **Context**   | `#87CEEB` (skyblue)       | `#002244` (deep navy)       | Fresher blue, stands out more than lightblue |
| **Note**      | `#FFD700` (gold)          | `#000000` (black)           | Bright, highlights important details |
| **Warning**   | `#FFA07A` (lightsalmon)   | `#660000` (darkred)         | Draws attention, not too harsh |
| **Info**      | `#C0C0C0` (silver)        | `#000000` (black)           | Neutral, "additional context" feeling |
| **Success**   | `#98FB98` (palegreen)     | `#004400` (forestgreen)     | Positive, "done" feeling |
| **Error**     | `#FF6666` (light red)     | `#330000` (very dark red)   | More alarming than light pink |
| **Highlight** | `#FFFFE0` (lightyellow)   | `#333333` (darkgray)        | Emphasizes key points |
