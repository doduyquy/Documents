# Remove local change(error) and get the lastest

- Sửa đổi vài thứ ở local nhưng không chạy, lỗi và muốn xóa toàn bộ, pull mới nhất về

## 1. Check current status and branch

To ensure: *main* or *dev*,...

```python
git status
git branch
```

## 2. Remove all local change(include uncommitted file)

```python
git restore .
```

- Or you can remove all file which is not tracked by Git

```python
git clean -fd
git restore .
```

## 3. Delete all local commit

```python
git fetch origin
git reset --hard origin/main
```

## 4. Update (sync 100%)

```python
git pull
```

## Summarize

| Trường hợp                   | Lệnh dùng                                          |
| ---------------------------- | -------------------------------------------------- |
| Chỉ sửa file chưa commit     | `git restore .`                                    |
| Có thêm/xóa file chưa commit | `git clean -fd && git restore .`                   |
| Có commit nhưng chưa push    | `git fetch origin && git reset --hard origin/main` |


> If you want to keep current code, use `stash`
